```python

import numpy as np
import matplotlib.pyplot as plt
arr = ["sky_chair", "ferris_wheel", "lobster_claw", "lobster_junior"]

counts = []
for i in arr:
  counts.append(park_data[i].value_counts()[1])

plt.bar(arr, counts, color ='maroon',width = 0.4)

plt.xlabel("Types of Rides")
plt.ylabel("visitors")
plt.title("Different rides taken by visitors")
plt.show()
```
